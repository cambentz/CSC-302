Please read the first 1000 lines from the lego_setsHB.csv in the DATA folder and split each line by using comma.
As you do go through each line populate two dictionaries, Price and Difficulty, which use product id as the key and price and review difficulty as values, respectively.


```python
# Read file, separate by line
file = open('/content/drive/MyDrive/DATA/lego_setsHB.csv', 'r')
lines = file.readlines()

# Initialize dictionaries
price = {}
difficulty = {}

# Check column headers
#print(lines[0])

# Read first 1000 lines, ignoring headers
for line in lines[1:1000]:

  # Split line values by comma
  parts = line.strip().split(',')

  # Save key value as variable
  product_id = parts[4]

  # Insert price and difficulty values into dictionaries by key value
  price[product_id] = parts[0]
  difficulty[product_id] = parts[5]

# Print data by key value
for product_id, price in price.items():
  print(product_id, price, difficulty[product_id])
```
